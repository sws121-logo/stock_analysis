```python
# moving average of various stocks
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
df['Mavg']=df['Close'].rolling(window=3).mean()
df.head(30)
```

```python
# daily return average of a stock= Closed Price - Open Price
df['Daily_returns']=df['Close']-df['Open']
# df['Daily_returns']=df['Adj Close'].pct_change()
df
```

```python
import numpy as np
def trend(x):
  if x > -0.5 and x <= 0.5:
    return 'Slight' or 'No change'
  elif x > 0.5 and x <= 1:
    return 'Slight Positive'
  elif x > -1 and x <= -0.5:
    return 'Slight Negative'
  elif x > 1 and x <= 3:
    return 'Positive'
  elif x > -3 and x <= -1:
    return 'Negative'
  elif x > 3 and x <= 7:
    return 'Among top gainers'
  elif x > -7 and x <= -3:
    return 'Among top losers'
  elif x > 7:
    return 'Bull run'
  elif x <= -7:
    return 'Bear drop'
df['Trend']= np.zeros(df['Daily_returns'].count())
df['Trend']= df['Daily_returns'].apply(lambda x:trend(x))
df.head(30)
```

```python
# Visualize trend frequency through a Pie Chart
import matplotlib.pyplot as plt
pie_data = df.groupby('Trend')
pie_label = sorted([i for i in df.loc[:, 'Trend'].unique()])
plt.pie(pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 1)
plt.legend(loc='upper left',fontsize=6)

plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Set the start and end date
start_date = '2021-01-01'
end_date = '2022-07-12'

# Define the ticker list
tickers_list = ['GOOGL', 'MSFT', 'AMZN']

# Create placeholder for data
data = pd.DataFrame(columns=tickers_list)

# Fetch the data
for ticker in tickers_list:
    data[ticker] = yf.download(ticker, 
                               start_date,
                               end_date)['Close']
    
# Print first 30 rows of the data
data.head(30)
```

```python
# correlation b/w different stock data
corr_df=data.corr()
corr_df
```

```python
#  correlate all stock data on heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(9,8))
sns.heatmap(corr_df,annot=True,cmap='RdYlGn')
plt.figure()
```
